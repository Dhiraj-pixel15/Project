```python
sns.distplot(Auto["Volume"])
plt.show()
```

```python
#Boxplot to check outliers

sns.boxplot(Auto["Volume"])
plt.show()
```

```python
Q1=np.quantile(Auto["Volume"],0.25)
Q2=np.quantile(Auto["Volume"],0.5)
Q3=np.quantile(Auto["Volume"],0.75)
```

```python
print("Q1:",Q1)
print("Q2:",Q2)
print("Q3:",Q3)
```

```python
IQR=Q3-Q1
print("IQR:",IQR)
```

```python
UW=Q3+(1.5*IQR)
LW=Q1-(1.5*IQR)

print("Upper Whisker:",UW)
print("Lower Whisker:",LW)
```
